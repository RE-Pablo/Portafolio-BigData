```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
regular = df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]]
regular
```

```python
sns.histplot(data = regular["SepalLengthCm"], kde=True)
plt.title("Histograma")
plt.show()
```

```python
sns.histplot(data = regular["SepalWidthCm"], kde=True)
plt.title("Histograma")
plt.show()
```

```python
sns.histplot(data = regular["PetalLengthCm"], kde=True)
plt.title("Histograma")
plt.show()
```

```python
sns.histplot(data = regular["PetalWidthCm"], kde=True)
plt.title("Histograma")
plt.show()
```

```python
sns.kdeplot(data = regular, x="SepalLengthCm", hue="Species", fill=True)
```

```python
sns.kdeplot(data = regular, x="SepalWidthCm", hue="Species", fill=True)
```

```python
sns.kdeplot(data = regular, x="PetalLengthCm", hue="Species", fill=True)
```

```python
sns.kdeplot(data = regular, x="PetalWidthCm", hue="Species", fill=True)
```
