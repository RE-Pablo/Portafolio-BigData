```python
import matplotlib.pyplot as plt 
```

```python
x = [x for x in range(2016,2024)]
x 
```

```python
import random 

y = [random.randint(100,200) for x in range (len(x))]
y2 = [random.randint(100,200) for x in range (len(x))]
y
```

Line Plot

```python
plt.figure(figsize=(6,6))

plt.plot(x,y, marker = "x", linestyle = "-.", color= "#D45f20")
plt.plot(x,y2, marker = "x", linestyle = "-", color= "b")
plt.title("Población de x país durante 2016-2023")
plt.xlabel("Año")
plt.ylabel("Población (M)")
plt.yticks([100,150,200])

plt.savefig("")
plt.show
```

Bar Plot

```python
x = ["México", "España","Brasil"]
y=[100,88,93]

plt.bar(x,y, color="y")
plt.show
```

Pie Chart

```python
plt.pie(y, labels = x)
plt.show
```

Histogram

```python
import random 
x = [random.randint(50,100) for x in range(32)]
bins=[50,60,70,80,90,100]

plt.hist(x,bins, edgecolor="k")
plt.show()
```

Box Plot- Diagrama de cajas y bigotes

```python
plt.boxplot(x)
plt.show()
```

Scatter plot- Gráfica de dispersión

```python
x = [random.randint(0,9) for x in range(30)]
y = [random.randint(0,9) for x in range(30)]

plt.scatter(x,[x*2 for x in range(len(x))])
plt.show()
```

Subplots

```python
import random 
x=[x for x in range(1950,2024)]
y=[random.randint(100,200) for x in range(len(x))]
y2=[random.randint(100,200)for x in range(len(x))]

fig, ax = plt.subplots(1,2)
ax[0].plot(x,y, color="g")
ax[1].plot(x,y2, color="r")
plt.show
```
